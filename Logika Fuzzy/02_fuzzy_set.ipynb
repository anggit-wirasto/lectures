Lima Soal Perhitungan Fungsi Keanggotaan Fuzzy

Soal 1: Fungsi Keanggotaan Segitiga
Diketahui fungsi keanggotaan fuzzy berbentuk segitiga:
    \mu_A(x) =
    \begin{cases} 
    0, & x \leq a \\
    \frac{x - a}{b - a}, & a < x \leq b \\
    \frac{c - x}{c - b}, & b < x \leq c \\
    0, & x \geq c
    \end{cases}
Dengan parameter a = 1, b = 4, dan c = 7.
Hitung nilai keanggotaan untuk x = 2, x = 5, dan x = 8.

Soal 2: Fungsi Keanggotaan Trapesium
Diketahui fungsi keanggotaan fuzzy berbentuk trapesium:
    \mu_A(x) =
    \begin{cases} 
    0, & x \leq a \\
    \frac{x - a}{b - a}, & a < x \leq b \\
    1, & b < x \leq c \\
    \frac{d - x}{d - c}, & c < x \leq d \\
    0, & x \geq d
    \end{cases}
Dengan parameter a = 2, b = 5, c = 7, dan d = 9.
Hitung nilai keanggotaan untuk x = 3, x = 6, dan x = 8.

Soal 3: Fungsi Keanggotaan Gaussian
Diketahui fungsi keanggotaan fuzzy berbentuk Gaussian:
    \mu_A(x) = e^{-\frac{(x - c)^2}{2\sigma^2}}
Dengan parameter c = 6 dan \sigma = 1.5.
Hitung nilai keanggotaan untuk x = 4 dan x = 7.

Soal 4: Fungsi Keanggotaan Sigmoid
Diketahui fungsi keanggotaan fuzzy berbentuk sigmoid:
    \mu_A(x) = \frac{1}{1 + e^{-k(x - c)}}
Dengan parameter k = 2 dan c = 5.
Hitung nilai keanggotaan untuk x = 3 dan x = 6.

Soal 5: Fungsi Keanggotaan Generalized Bell
Diketahui fungsi keanggotaan fuzzy berbentuk generalized bell:
    \mu_A(x) = \frac{1}{1 + \left| \frac{x - c}{a} \right|^{2b}}
Dengan parameter a = 2, b = 3, dan c = 6.
Hitung nilai keanggotaan untuk x = 3 dan x = 8.


```python

import numpy as np
import matplotlib.pyplot as plt

def triangular(x, a, b, c):
    return np.where(
        (x <= a) | (x >= c), 0,
        np.where(x <= b, (x - a) / (b - a), (c - x) / (c - b))
    )

def trapezoidal(x, a, b, c, d):
    return np.where(
        (x <= a) | (x >= d), 0,
        np.where(x <= b, (x - a) / (b - a), np.where(x <= c, 1, (d - x) / (d - c)))
    )

def gaussian(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

def sigmoid(x, k, c):
    return 1 / (1 + np.exp(-k * (x - c)))

def generalized_bell(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Sample points
x_vals = np.linspace(0, 10, 100)

# Compute membership functions
tri_vals = triangular(x_vals, 1, 4, 7)
trap_vals = trapezoidal(x_vals, 2, 5, 7, 9)
gauss_vals = gaussian(x_vals, 6, 1.5)
sigmoid_vals = sigmoid(x_vals, 2, 5)
gbell_vals = generalized_bell(x_vals, 2, 3, 6)

# Plot the functions
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0, 0].plot(x_vals, tri_vals, label="Triangular")
axs[0, 0].set_title("Triangular Membership Function")
axs[0, 1].plot(x_vals, trap_vals, label="Trapezoidal", color='r')
axs[0, 1].set_title("Trapezoidal Membership Function")
axs[0, 2].plot(x_vals, gauss_vals, label="Gaussian", color='g')
axs[0, 2].set_title("Gaussian Membership Function")
axs[1, 0].plot(x_vals, sigmoid_vals, label="Sigmoid", color='m')
axs[1, 0].set_title("Sigmoid Membership Function")
axs[1, 1].plot(x_vals, gbell_vals, label="Generalized Bell", color='c')
axs[1, 1].set_title("Generalized Bell Membership Function")

# Hide unused subplot
axs[1, 2].axis('off')

# Show plots
for ax in axs.flat:
    ax.legend()
    ax.grid()
plt.tight_layout()
plt.show()

# Compute and display specific values
x_test_values = [2, 5, 8, 3, 6, 7, 4]
results = {
    'Triangular': [triangular(x, 1, 4, 7) for x in x_test_values],
    'Trapezoidal': [trapezoidal(x, 2, 5, 7, 9) for x in x_test_values],
    'Gaussian': [gaussian(x, 6, 1.5) for x in x_test_values],
    'Sigmoid': [sigmoid(x, 2, 5) for x in x_test_values],
    'Generalized Bell': [generalized_bell(x, 2, 3, 6) for x in x_test_values]
}

for key, values in results.items():
    print(f"{key} Membership Values: {values}")
```
